```python
# Your code goes here
import pandas as pd


#breakline
def breakline():
  return print('\n','\n')
# Loading the datasets

# First file
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df1 = pd.read_csv(url1)
print('df1: ')
display(df1.head())
print(df1.shape)
breakline()


# Second file
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url2)
print('df2: ')
display(df2.head())
print(df2.shape)
breakline()


# Third file
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url3)
print('df3: ')
display(df3.head())
print(df3.shape)
breakline()
```

```python
# Preparing the columns
df1.columns = list(map(lambda x: x.lower().replace(" ","_").strip(), df1.columns))
df1= df1.rename({'st' : 'state'}, axis = 1)
display(df1.columns)

df2.columns = list(map(lambda x: x.lower().replace(" ","_").strip(), df2.columns))
df2= df2.rename({'st' : 'state'}, axis = 1)
display(df2.columns)

df3.columns = list(map(lambda x: x.lower().replace(" ","_").strip(), df3.columns))
display(df3.columns)
```

```python
# Concatenating the 3 dataframes
df = pd.concat([df1, df2, df3])
df
```

```python
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
df.duplicated().value_counts()
```

```python
df_test = df.dropna(axis = 0, how ="all")
df_test
```

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
df.head()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df.pivot_table(values='total_claim_amount', index='sales_channel', aggfunc='sum').round(2)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Pivot table with average income per gender and education level
pivot_table = pd.pivot_table(df, values='customer_lifetime_value', index=['gender', 'education'], aggfunc='mean')

pivot_table
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Create a pivot table with the number of open complaints by month and policy type
pivot_table = pd.pivot_table(df, values='number_of_open_complaints', index=['month'], columns=['policy_type'], aggfunc='count', fill_value=0)
pivot_table
```

```python
# Stack the pivot table to create a long format table
pivot_table.stack().reset_index()
```
